```python
import pandas as pd

train_info_rt = '../../data/train_info.txt'
test_info_rt = '../../data/test_info.txt'
user_info_rt = '../../data/user_info.txt'
doc_info_rt = '../../data/news_info.txt'

train_raw = pd.read_csv(train_info_rt, sep='\t', header=None)
test_raw = pd.read_csv(test_info_rt, sep='\t', header=None)
user_info_raw = pd.read_csv(user_info_rt, sep='\t', header=None)
doc_info_raw = pd.read_csv(doc_info_rt, sep='\t', header=None)
```

```python
train_raw.columns = ['userid', 'docid', 'timestamp', 'network', 'refresh', 'position', 'click', 'duration']
test_raw.columns = ['id', 'userid', 'docid', 'timestamp', 'network', 'refresh']
user_info_raw.columns = ['userid', 'device', 'os', 'province', 'city', 'age', 'gender']
doc_info_raw.columns = ['docid', 'title', 'pubtime', 'picnum', 'category1st', 'category2nd', 'keyword']
```

```python
user_id = list(set(list(user_info_raw['userid'])))
user_id.sort()
user_id_new_index = list(range(1, len(user_id) + 1))
user_id_dict = dict(zip(user_id, user_id_new_index))
user_info_raw['userid'] = user_info_raw['userid'].map(user_id_dict)
train_raw['userid'] = train_raw['userid'].map(user_id_dict)
test_raw['userid'] = test_raw['userid'].map(user_id_dict)
```

```python
doc_id = list(set(list(doc_info_raw['docid'])))
doc_id.sort()
doc_id_new_index = list(range(1, len(doc_id) + 1))
doc_id_dict = dict(zip(doc_id, doc_id_new_index))
doc_info_raw['docid'] = doc_info_raw['docid'].map(doc_id_dict)
train_raw['docid'] = train_raw['docid'].map(doc_id_dict)
test_raw['docid'] = test_raw['docid'].map(doc_id_dict)
```

```python
# 导出数据
train_raw.to_csv('../../data/train_info_new_id.txt', sep='\t', header=None, index=False)
test_raw.to_csv('../../data/test_info_new_id.txt', sep='\t', header=None, index=False)
user_info_raw.to_csv('../../data/user_info_new_id.txt', sep='\t', header=None, index=False)
doc_info_raw.to_csv('../../data/news_info_new_id.txt', sep='\t', header=None, index=False)
```
