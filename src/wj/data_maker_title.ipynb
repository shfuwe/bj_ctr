```python
import jieba
import numpy as np
def tokenize(text,stop):
    """
    带有语料清洗功能的分词函数
    """
#     text = re.sub("\{%.+?%\}", " ", text)           # 去除 {%xxx%} (地理定位, 微博话题等)
#     text = re.sub("@.+?( |$)", " ", text)           # 去除 @xxx (用户名)
#     text = re.sub("【.+?】", " ", text)              # 去除 【xx】 (里面的内容通常都不是用户自己写的)
#     icons = re.findall("\[.+?\]", text)             # 提取出所有表情图标
#     text = re.sub("\[.+?\]", "IconMark", text)      # 将文本中的图标替换为`IconMark`

    tokens = []
    s=''
    for k, w in enumerate(jieba.lcut(text)):
        w = w.strip()#移除字符串头尾指定的字符
        if w in stop:
            continue
        if w.encode('utf-8').isalnum() is True:
          # print(w,w.encode('utf-8').isalnum())
          continue
        tokens.append(w)
        s+=w
        s+=" "
    return tokens,s
stopwords = []
with open("stop_wb.txt", "r", encoding="utf8") as f:
    for w in f:
        stopwords.append(w.strip())

def get_clear(text):
    text_clear=[]
    for i in text:
        if isinstance(i, str):
            tokens,s=tokenize(i,stopwords)
            text_clear.append(tokens)
    return text_clear

def sen_v(text,stopwords):
    try:
        tokens,s=tokenize(text,stopwords)
    except:
        print('tokenize error')
        print(text)
        tokens=[]
        
    all_=[]
    for i in tokens:
        try:
#             print(model0.wv(i))
            all_.append(model0.wv[i])
#             print(all_)
        except:
            continue
    
    all_=np.array(all_)
    
    if len(all_)>0:
        all_=np.average(all_, axis=0)
    else:
        print('all_ error np.zeros(300)')
        print(text)
        all_=np.zeros(300)
    return all_

from gensim.models import Word2Vec
model0 = Word2Vec.load('v300.model')
```
