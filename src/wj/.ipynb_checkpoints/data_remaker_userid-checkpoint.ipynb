```python
import pandas as pd
def store_list(lsit,text_path):
    ff = open(text_path, encoding='utf-8', mode='w')
    for line_list in lsit:
        ff.write(str(line_list))  # 写入一个新文件中
        ff.write("\n")
```

```python
def read_list(text_path):
    lsit=[]
    with open('%s' % text_path, 'r', encoding="utf8") as f:  # 打开一个文件只读模式
        line = f.readlines()  # 读取文件中的每一行，放入line列表中
        for line_list in line:
            lsit.append(int(line_list.replace('\n','')))
    return lsit
```

```python
import time
import datetime
# 定义日志（data文件夹下，同级目录新建一个data文件夹）
def write_log(w):
    file_name = '../../data/wj/' + datetime.date.today().strftime('%m%d') + "_{}.log".format("data_remaker_userid")
    t0 = datetime.datetime.now().strftime('%H:%M:%S')
    info = "{} : {}".format(t0, w)
    print(info)
    with open(file_name, 'a') as f:
        f.write(info + '\n')
```

```python
train_info = pd.read_table('../../data/train_info_new_id.txt',header=None)
train_info.columns = ['userid', 'docid', 'timestamp', 'network', 'refresh', 'position', 'click', 'duration']
```

```python
doc_info = pd.read_pickle('../../data/wj/doc.pkl')
#造userid feature
def feature_maker(biao,s): 
    agg_dict1 = {'click':['mean','count'], 'duration':['mean'], 'picnum':['mean']}
    made = biao.groupby([s]).agg(agg_dict1)
    made=made.reset_index()
    made.columns=[s,s+'_click_mean',s+'_click_count',s+'_duration_mean',s+'_picnum_mean']
    
    for i in made.columns:
        if i==s:
            continue
        
        made[i] = made[[i]].apply(max_min_scaler)
        
    return made
# 归一化函数
max_min_scaler = lambda x: (x - np.min(x)) / (np.max(x) - np.min(x))
```

```python
train_info_doc=pd.merge(train_info,doc_info,on='docid')
train_info_doc
```

```python
import numpy as np
userid_f=feature_maker(train_info_doc,'userid')
userid_f
```

```python
userid_click_mean=[]
userid_click_count=[]
userid_duration_mean=[]
userid_picnum_mean=[]

from tqdm import tqdm
for i in tqdm(range(len(nb_df))):
    userid_click_mean2=0
    userid_click_count2=0
    userid_duration_mean2=0
    userid_picnum_mean2=0
    
#     nb_id=nb_df.loc[i]['userid'].item()
    nb_close=nb_df.loc[i]['close_id']
    for tid in nb_close:
        tid_=userid_f.loc[userid_f['userid']==tid]
        
        userid_click_mean2+=tid_['userid_click_mean'].item()
        userid_click_count2+=tid_['userid_click_count'].item()
        userid_duration_mean2+=tid_['userid_duration_mean'].item()
        userid_picnum_mean2+=tid_['userid_picnum_mean'].item()
    
    userid_click_mean.append(userid_click_mean2/20)
    userid_click_count.append(userid_click_count2/20)
    userid_duration_mean.append(userid_duration_mean2/20)
    userid_picnum_mean.append(userid_picnum_mean2/20)

nb_df['userid_click_mean']=userid_click_mean
nb_df['userid_click_count']=userid_click_count
nb_df['userid_duration_mean']=userid_duration_mean
nb_df['userid_picnum_mean']=userid_picnum_mean
nb_df
```
