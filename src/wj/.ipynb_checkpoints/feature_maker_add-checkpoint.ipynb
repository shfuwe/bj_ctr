```python
import pandas as pd
df_train = pd.read_table('../../data/train_info_new_id.txt',header=None)
df_train.columns = ['userid', 'docid', 'timestamp', 'network', 'refresh', 'position', 'click', 'duration']
```

```python
for i in ['docid', 'timestamp','position']:
    if i in df_train.columns:
        del df_train[i]
df_train.head()
```

```python
user_info = pd.read_pickle('../../data/wj/user.pkl')
user_info.head()
```

```python
train_info_doc=pd.merge(df_train,user_info,on='userid')
train_info_doc
```

```python
train_info_user=train_info_doc
```

```python
# 归一化函数
import numpy as np
max_min_scaler = lambda x: (x - np.min(x)) / (np.max(x) - np.min(x))
def feature_maker(biao,s): 
    agg_dict1 = {'click':['mean','count'], 'duration':['mean']}
    made = biao.groupby([s]).agg(agg_dict1)
    made=made.reset_index()
    made.columns=[s,s+'_click_mean',s+'_click_count',s+'_duration_mean']
    
    for i in made.columns:
        if i==s:
            continue
        
        made[i] = made[[i]].apply(max_min_scaler)
        
    return made
```

```python
network_f=feature_maker(train_info_user,'network')
network_f
```

```python
refresh_f=feature_maker(train_info_user,'refresh')
refresh_f
```

```python
device_f=feature_maker(train_info_user,'device')
device_f
```

```python
os_f=feature_maker(train_info_user,'os')
os_f
```

```python
province_f=feature_maker(train_info_user,'province')
province_f
```

```python
city_f=feature_maker(train_info_user,'city')
city_f
```

```python
age_f=feature_maker(train_info_user,'age')
age_f
```

```python
gender_f=feature_maker(train_info_user,'gender')
gender_f
```

```python
def fill_(pdd,s):
    mean=pdd[s].mean()
    pdd[s].fillna(mean, inplace=True)
    return pdd
```

```python
from tqdm import tqdm
```

```python
s_l=[]
_f=[]
ff=['network','refresh','device','os','province','city','age','gender']
for i in ff:
    s_l.append(i+'_click_mean')
    s_l.append(i+'_click_count')
    s_l.append(i+'_duration_mean')
    _f.append(i+'_f')
print(_f)
print(s_l)
```

```python
def maker_add(df):
    print(df.shape)
    for i in range(len(ff)):
        name=ff[i]
        table_name=_f[i]
        df=pd.merge(df,globals()[table_name],how='left',on=name)
        print(df.shape)
    from tqdm import tqdm
    for s in tqdm(s_l):
        df=fill_(df,s)
    return df
    
```

```python
import pandas as pd
df_test_user_doc=pd.read_pickle('../../data/wj/df_test_user_doc.pkl')
df_test_user_doc=maker_add(df_test_user_doc)
df_test_user_doc.head()
```

```python
df_test_user_doc.to_pickle('../../data/wj/df_test_user_doc_64.pkl')
```

```python
def nbnb(path)
    df_one=pd.read_pickle(path+'.pkl')
    df_one=maker_add(df_one)
    df_one.to_pickle(path+'_64.pkl')
```
