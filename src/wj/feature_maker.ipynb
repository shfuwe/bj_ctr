```python
#造userid feature
def feature_maker(biao,s): 
    agg_dict1 = {'click':['mean','count'], 'duration':['mean'], 'picnum':['mean']}
    made = biao.groupby([s]).agg(agg_dict1)
    made=made.reset_index()
    made.columns=[s,s+'_click_mean',s+'_click_count',s+'_duration_mean',s+'_picnum_mean']
    
    for i in made.columns:
        if i==s:
            continue
        
        made[i] = made[[i]].apply(max_min_scaler)
        
    return made
```

```python
s_l=['userid_click_mean','userid_click_count' ,'userid_duration_mean' ,'userid_picnum_mean',
'docid_click_mean','docid_click_count','docid_duration_mean','docid_picnum_mean',
'category1st_click_mean','category1st_click_count','category1st_duration_mean','category1st_picnum_mean',
'category2nd_click_mean','category2nd_click_count','category2nd_duration_mean','category2nd_picnum_mean',
'keyword0_click_mean','keyword0_click_count','keyword0_duration_mean','keyword0_picnum_mean']
from tqdm import tqdm
```

```python
def fill_(pdd,s):
    mean=pdd[s].mean()
    pdd[s].fillna(mean, inplace=True)
    return pdd
```

```python
df_test_user=pd.merge(df_test,user_info,on='userid')
df_test_user_doc=pd.merge(df_test_user,doc_info,on='docid')
print(df_test_user_doc.shape)

df_test_user_doc=pd.merge(df_test_user_doc,userid_f,how='left',on='userid')
print(df_test_user_doc.shape)

df_test_user_doc=pd.merge(df_test_user_doc,doc_f,how='left',on='docid')
print(df_test_user_doc.shape)

df_test_user_doc=pd.merge(df_test_user_doc,category1st_f,how='left',on='category1st')
print(df_test_user_doc.shape)

df_test_user_doc=pd.merge(df_test_user_doc,category2nd_f,how='left',on='category2nd')
print(df_test_user_doc.shape)

df_test_user_doc=pd.merge(df_test_user_doc,keyword0_f,how='left',on='keyword0')
print(df_test_user_doc.shape)

for s in tqdm(s_l):
    df_test_user_doc=fill_(df_test_user_doc,s)
    
print(df_test_user_doc.shape)
print(df_test_user_doc)

df_test_user_doc.sort_values("id",inplace=True)
df_test_user_doc.index = range(len(df_test_user_doc))
del df_test_user_doc['id']
```
