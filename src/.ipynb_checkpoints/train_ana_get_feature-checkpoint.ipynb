```python
import pandas as pd
train_info = pd.read_table('../data/train_info.txt',header=None)
print(train_info)
train_info.columns = ['userid','docid','timestamp','network','refresh','position','click','duration']
print(train_info)
```

```python
del train_info['timestamp']
del train_info['network']
del train_info['refresh']
del train_info['position']
print(train_info)
```

```python
a=train_info.groupby(['userid','docid']).sum()
```

```python
b=a.reset_index()
print(b)
```

```python
b.columns = ['userid','docid','userid_docid_click','userid_docid_duration']
print(b)
```

```python
b.to_pickle('../data/userid_docid_ana.pkl')
```

```python
print(userid)
```

```python
# del userid['docid']
userid=userid.reset_index()
```

```python
doc_info = pd.read_pickle('../data/out1/doc_info1.pkl')
print(doc_info)
```

```python
train_info_doc=pd.merge(train_info,doc_info,on='docid')
print(train_info_doc)
```

```python
del train_info_doc['keyword1']
del train_info_doc['keyword2']
del train_info_doc['keyword3']
del train_info_doc['keyword4']
del train_info_doc['pubtime']
print(train_info_doc)
```

```python
# data9 = train_info_doc.groupby(['userid'])[['click','duration','picnum']].agg(['mean', 'count'])
```

```python
agg_dict1 = {'click':['mean','count'], 'duration':['mean'], 'picnum':['mean']}
user = train_info_doc.groupby(['userid']).agg(agg_dict1)
```

```python
userid=user.reset_index()
print(userid)
```

```python
userid.columns = ['userid','userid_click_mean','userid_click_count','userid_duration_mean','userid_picnum_mean']
print(userid)
```

```python
userid.to_pickle('../data/userid_.pkl')
```
